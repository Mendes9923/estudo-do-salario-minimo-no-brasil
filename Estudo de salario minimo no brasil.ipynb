```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Dados fictícios
anos = list(range(2000, 2024))
salario_nominal = [200 + 20*i for i in range(24)]
inflacao_acumulada = [1.02**i for i in range(24)]
salario_real = [salario_nominal[i] / inflacao_acumulada[i] for i in range(24)]
cesta_basica = [300 + 15*i for i in range(24)]
taxa_emprego = [95 - 0.2*i for i in range(24)]
```

```python
# DataFrame para facilitar a manipulação dos dados
df = pd.DataFrame({
    'Ano': anos,
    'Salário Nominal': salario_nominal,
    'Salário Real': salario_real,
    'Cesta Básica': cesta_basica,
    'Taxa de Emprego': taxa_emprego
})
```

```python
# Gráfico 1: Evolução do Salário Mínimo Nominal e Real
plt.figure(figsize=(10, 5))
plt.plot(df['Ano'], df['Salário Nominal'], label='Salário Nominal', marker='o')
plt.plot(df['Ano'], df['Salário Real'], label='Salário Real (Ajustado pela Inflação)', marker='o')
plt.title('Evolução do Salário Mínimo Nominal e Real (2000-2023)')
plt.xlabel('Ano')
plt.ylabel('Valor (R$)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gráfico 2: Poder de Compra do Salário Mínimo
plt.figure(figsize=(10, 5))
plt.plot(df['Ano'], df['Salário Real'], label='Salário Real (Ajustado pela Inflação)', marker='o')
plt.plot(df['Ano'], df['Cesta Básica'], label='Custo da Cesta Básica', marker='o')
plt.title('Poder de Compra do Salário Mínimo (2000-2023)')
plt.xlabel('Ano')
plt.ylabel('Valor (R$)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gráfico 3: Impacto do Salário Mínimo no Emprego
plt.figure(figsize=(10, 5))
plt.plot(df['Ano'], df['Taxa de Emprego'], label='Taxa de Emprego (%)', marker='o', color='green')
plt.title('Impacto do Salário Mínimo no Emprego (2000-2023)')
plt.xlabel('Ano')
plt.ylabel('Taxa de Emprego (%)')
plt.legend()
plt.grid(True)
plt.show()
```
